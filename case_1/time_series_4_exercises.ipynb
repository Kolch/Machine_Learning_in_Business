```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd 
import seaborn as sns
import statsmodels.api as sm
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import mean_squared_error
```

```python
def show_feature_importances(feature_names, feature_importances, get_top=None):
    feature_importances = pd.DataFrame({'feature': feature_names, 'importance': feature_importances})
    feature_importances = feature_importances.sort_values('importance', ascending=False)
       
    plt.figure(figsize = (20, len(feature_importances) * 0.355))
    
    sns.barplot(feature_importances['importance'], feature_importances['feature'])
    
    plt.xlabel('Importance')
    plt.title('Importance of features')
    plt.show()
    
    if get_top is not None:
        return feature_importances['feature'][:get_top].tolist()
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
Path_Energy = 'my_PJME_MW.csv'
data = pd.read_csv(Path_Energy, index_col=[0], parse_dates=[0])
```

```python
df_w = data.resample('W').mean()
```

```python
df_w.head()
```

```python
sm.graphics.tsa.plot_acf(df_w['PJME_MW'].values, lags=52)
plt.show()

sm.graphics.tsa.plot_pacf(df_w['PJME_MW'].values, lags=52)
plt.show()
```

```python
# добавим историю на глубину в 5 дней
p = 5
for i in range(p):
    df_w['t'+str(p-i)]=0
    df_w['t'+str(p-i)].values[p:] = df_w['PJME_MW'].values[i:-(p)+i]
```

```python
first_diff =np.array(df_w.iloc[1:,0].values - df_w.iloc[:-1,0].values).astype('float32') 
df_w['diff1'] = 0
df_w['diff1'] =  df_w['diff1'].astype('float') 
df_w['diff1'].values[1:] = first_diff
```

```python
# добавим разности на глубину 3
p = 5
for i in range(p):
    df_w['d'+str(p-i)]=0
    df_w['d'+str(p-i)].values[p:] = df_w['diff1'].values[i:-(p)+i]
```

```python
def exponential_smoothing(series, alpha):
    result = [series[0]] 
    for n in range(1, len(series)):
        result.append(alpha * series[n] + (1 - alpha) * result[n-1])
    return result
```

```python
df_w['exp_0_01'] = exponential_smoothing(df_w['PJME_MW'], 0.01)
df_w['exp_0_03'] = exponential_smoothing(df_w['PJME_MW'], 0.03)
df_w['exp_0_04'] = exponential_smoothing(df_w['PJME_MW'], 0.04)
df_w['exp_0_06'] = exponential_smoothing(df_w['PJME_MW'], 0.06)
df_w['exp_0_07'] = exponential_smoothing(df_w['PJME_MW'], 0.07)
df_w['exp_0_08'] = exponential_smoothing(df_w['PJME_MW'], 0.08)
```

### Отбор признаков

```python
h = 1
X = df_w.iloc[p:-h,:]
X.shape
```

```python
y = df_w['PJME_MW'].values[h+p:].reshape((df_w.shape[0]-h-p,1))
y.shape
```

```python
model_rfr = RandomForestRegressor(n_estimators=500, random_state=42, max_depth=20)
model_rfr.fit(X, y)
features = show_feature_importances(X.columns, model_rfr.feature_importances_, get_top=10)
```

### Деление данных на трейн и тест

```python
def split_data( data, split_date ):
    return data.loc[data.index.get_level_values('Date') <= split_date].copy(), \
           data.loc[data.index.get_level_values('Date') >  split_date].copy()
```

```python
train, test = split_data(df_w[features], '01-07-2016')

X_train = train.iloc[:-1,:]
y_train = train['PJME_MW'].values[1:]
X_test = test.iloc[:-1,:]
y_test = test['PJME_MW'].values[1:]

X_train_4 = train.iloc[:-4,:]
y_train_4 = train['PJME_MW'].values[4:]
X_test_4 = test.iloc[:-4,:]
y_test_4 = test['PJME_MW'].values[4:]

plt.figure( figsize=( 15, 7 ))
plt.xlabel('Время')
plt.ylabel('Потребление энергии, МВт')
plt.plot(train.index.get_level_values('Date'),train['PJME_MW'], label='train data' )
plt.plot(test.index.get_level_values('Date'),test['PJME_MW'], label='test data')
plt.title( 'Тестовые и тренировочные данные' )
plt.legend()
plt.show()
```

### Построение модели

#### h = 1

```python
model_rfr = RandomForestRegressor(n_estimators=500, random_state=42, max_depth=20)
model_rfr.fit(X_train, y_train)
y_test_pred = model_rfr.predict(X_test)
```

#### h = 4

```python
model_rfr.fit(X_train_4, y_train_4)
y_test_pred_4 = model_rfr.predict(X_test_4)
```

```python
plt.figure( figsize=( 18, 10))


plt.xlabel('Время')
plt.ylabel('Потребление энергии, МВт')
plt.plot(test.index[1:],y_test, label='test data')
plt.plot(test.index[1:],y_test_pred, 'g', label='predict data h = 1')
plt.plot(test.index[4:],y_test_pred_4, '--r', label='predict data h = 4')

plt.legend()
plt.show()

er_g1 = mean_absolute_percentage_error(y_true=y_test, y_pred=y_test_pred)
er_g4 = mean_absolute_percentage_error(y_true=y_test[3:], y_pred=y_test_pred_4)
print('ошибка h=1:  ', er_g1, '%')
print('ошибка h=4:  ', er_g4, '%')
```

Как видим, модель с h=1 работает лучше, чем модель h=4. Тем не менее модель с h=4 все же понимает сезонность и с ошибками, но следует за истинными данными. Ошибка состовляет 6%, думаю можно сделать намного лучше, это вопрос работы с признаками.
